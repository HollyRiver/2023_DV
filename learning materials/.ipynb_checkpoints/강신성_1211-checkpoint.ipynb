> 강의소개? (Quarto 대시보드 만들기)

## 1. 강의소개?

`-` 데이터시각화

* 통계학과(빅데이터 등)
> 탐색적 자료분석, eda
>
> 예시 : 성별 타이타닉 생존자 수 시각화
>
> 목표 : 데이터 해석 및 분석(해석을 잘 해야 한다. 상관관계/인과관계 차이, 신뢰구간/잔차분석/QQplot/히트맵 등등...)
>
> 특징 : 그림을 그려놓고 해석을 함, 유의미한 메시지를 읽고자 함 + 통계적 처리를 해서 그림을 그림
>
> 툴 : R, Python
>
> 필수요소 : 그림을 그리는 방법을 알아야 함 + 그림을 해석할 줄 알아야 함. 코딩 + 데이터 이해
>
> 수업 범위 : R/Python 문법 리뷰 --> EDA --> 지리정보/텍스트자료 등 특이한 자료 시각화
>
> 잘한다 소리 들으려면... : 데이터를 보고 숨어있는 정보를 도출해낼 수 있어야 함.

**데이터시각화는 통계학과에서만 관심이 있냐?? no...**

* 컴공
> 웹, 플랫폼(크롤링 + 시각화 --> 플랫폼, 시각화 결과를 웹으로 배포)
>
> 예시 : 코로나 19 대시보드 (통계분석이 그다지 중요하지 않음.)
>
> 목표 : 시각화 플랫폼 개발 / 시각화 앱 개발 / 대시보드 설계 및 배포 --> 내가 만든 시각화를 웹에 배포하여 타인이 쉽게 접근할 수 있도록 하는 것.
>
> 특징 : 인터랙티브 그래프, 움직이는 시각화를 선호
>
> 툴 : js
>
> 필수요소 : 코딩능력 + 배포
>
> 수업 범위 : 여러가지 문법 리뷰 --> 웹서버의 개념 --> 대시보드(프로젝트)
>
> 잘한다 소리 들으려면... : 우와 막 움직인다

* 시각디자인
> 인포그래픽, 시각적 기법(팜플렛 느낌?)
>
> 예시 : 커피 소비량 인포그래픽
>
> 목표 : 인상적이고 기억에 남는 그림, 작품 제작
>
> 특징 : 아기자기하고 예쁨, 다채로움
>
> 툴 : 일러스트레이터
>
> 필수요소 : 그래픽 툴 사용능력 + 미적 감각
>
> 수업 범위 : 색 이론(색의 배합, 보색 등) + 인지이론(3D 플랏의 인지 왜곡 등)
>
> 잘한다 : 그림이 예쁘다, 야무지다.


* 학원(패스트 캠퍼스)
> 키워드 : 공모전, 실무
>
> 예시 : 데이터 시각화 공모전
>
> 목표 : 취업 스펙, 실무 기술 취득
>
> 특징 : 배우기 쉬움 (해석 X, 코딩 X, 디자인 툴 X) + 엑셀 상위호환 느낌
> 
> 툴 : 태블로, PowerBI
>
> 필수요소 : 툴 사용 능력
>
> 수업 범위 : 툴 사용 능력 배양
>
> 수업을 안하는 이유 : 유료(회사에 따라서 이걸 배우면 불리할 수도 있음) + 대학에서 다룰 내용에서 벗어나는 듯


`-` 우리 수업 목표

* 그림을 읽는 능력, 데이터를 올바르게 해석하는 능력 // 통계학과 주무기
* 파이썬에 대한 매우 높은 숙련도 // 이게 약하면 태블로에게 밀림
* 기본적인 미적교양(aes, 터프티와 나이젤홈즈, 훌륭한 시각화)
* 반응형 플랏, 대시보드, 지리정보 시각화 // 이게 약하면 IT계열에게 밀림

# 대시보드 만들기

```
project:
  type: website
  output-dir: docs
```

> 클론한 리포지토리에 `quarto.yml`을 만들고 위의 내용을 입력 후 저장.
