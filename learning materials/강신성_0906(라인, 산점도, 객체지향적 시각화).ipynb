###겹쳐 그리기

```python
plt.plot([1,2,3,2], '--o', color = 'orange')
plt.plot([2,3,1,4], '--o', color = 'skyblue')

plt.show()

plt.plot([4,4,2,1], '--o', color = 'cyan')
```

```python
plt.plot([1,2,3,2], '--o', color = 'C1')
plt.plot([2,3,1,4], '--o', color = 'C0')

plt.show()
```

```python
x = np.arange(-5,5,0.1)
eps = np.random.randn(100)
y = 2*x + eps ## 벗어나도록 겹치게
```

```python
plt.plot(x,y,'.b')
plt.plot(x,2*x,'--r')
plt.show()
```
