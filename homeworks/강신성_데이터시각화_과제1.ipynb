```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

2. 파일 불러오기 및 행렬로 저장

3. 색상별 행렬 재분류 및 명도 변환

```python
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]
```

```python
rr = cv2.equalizeHist(r)
gg = cv2.equalizeHist(g)
bb = cv2.equalizeHist(b)
```

```python
img_equalized = np.stack([rr,gg,bb], axis = -1)
```

4. 결과값 출력 및 비교

```python
plt.imshow(img_equalized);
```

```python
plt.imshow(np.concatenate([img, img_equalized], axis = 1));
```
