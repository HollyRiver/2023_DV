### **분할된 그래프 그리기**

**1. 사전작업**

> 라이브러리 설치

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

> 그래프 사이즈 및 dpi 설정

```python
matplotlib.rcParams['figure.figsize'] = (3, 2)
matplotlib.rcParams['figure.dpi'] = 75
```

**2. x와 y값 입력**



```python
x, y = [1,2,3,4], [1,2,4,3]
```

**3. 객체 설정 및 그래프 산출**

```python
fig, ((ax1, ax2, ax3),(ax4, ax5, ax6)) = plt.subplots(2,3, figsize = (5,3.5)) ## fig, axes 지정
ax1.plot(x, y, 'or')  ## ax1, marker = 'o', color = 'red'
ax2.plot(x, y, 'og')
ax3.plot(x, y, 'ob')
ax4.plot(x, y, '--or')  ## ax4, marker = 'o', linetype = '--', color = 'red'
ax5.plot(x, y, '--og')
ax6.plot(x, y, '--ob')

fig.tight_layout()
```
