### 1. 사전작업

* 라이브러리 설치

````python
#!pip install plotnine
````

````python
import pandas as pd
import numpy as np

from plotnine import *
import plotnine
````

* figure 사이즈 설정

````python
plotnine.options.dpi= 150
plotnine.options.figure_size = (6, 5)
````

* 데이터 가져오기

````python
df = pd.read_csv('https://raw.githubusercontent.com/guebin/DV2022/master/posts/mpg.csv')
df.head()
````

### 2. 데이터 전처리

````python
mpg = df.assign(cyl_cut = pd.cut(df.cyl, bins = 2))   ## cyl : 실린더 수에 해당하는 열을 두 층으로 분해한다.
mpg.head()
````

> 필요한 자료는 ```cyl```, ```displ```, ```cty```이다.

* x축에 ```displ```, y축에 ```cty```, 그리고 ```cyl```을 두 구간으로 나눈 ```cyl_cut```을 두 색상으로 구분하여 산출하여야 한다.

### 3. 그래프 산출

````python
fig = ggplot(mpg)
point = geom_point(aes(x = 'displ', y = 'cty', color = 'cyl_cut'))
smooth = geom_smooth(aes(x = 'displ', y = 'cty', color = 'cyl_cut'))
````

> 따라서 아래와 같은 결과를 산출시킨다.

````python
fig + point + smooth
````
