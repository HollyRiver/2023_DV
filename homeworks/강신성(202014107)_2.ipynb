**1. 라이브러리 설치**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**2. 벡터 지정**

```python
x = np.arange(-5,5,0.1)
y1 = np.sin(x)
y2 = np.sin(2*x) + 2
y3 = np.sin(4*x) + 4
y4 = np.sin(8*x) + 6
```

**3. 그래프 산출**

```python
plt.plot(x,y1,'--',color = '#ff3333')
plt.plot(x,y2,'--',color = '#128912')
plt.plot(x,y3,'--b')
plt.plot(x,y4,'--',color = '#ca2eca')

plt.show()
```
