```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/HollyRiver/2023_DV/main/dummy/df.csv')
```

```python
df.datetime = pd.to_datetime(df.datetime)
```

```python
sns.lineplot(
    df.sort_values('datetime')[:(24*28)], ## 28   
    x = 'datetime', y = 'count',
    hue = 'dataset_type', linestyle = '--',
    lw = 0.8
)

sns.lineplot(
    df.sort_values('datetime')[:(24*28)],
    x = 'datetime', y = 'count_hat',
    hue = 'dataset_type',
    lw = 3, alpha = 0.5
)

fig = plt.gcf()
fig.set_size_inches(8,2)  ##  
plt.xticks(rotation = 15);  ## x  
```
