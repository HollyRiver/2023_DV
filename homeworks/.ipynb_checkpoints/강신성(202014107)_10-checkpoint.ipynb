## 문제

적당한 데이터프레임을 선언하고 아래와 같은 barplot을 그려라.

![](https://github.com/HollyRiver/Data_Visualization/blob/84fedc269512a219c9665b6c6a6b414521af405d/homeworks/cell-24-output-1.png?raw=true)

## 풀이

### **A. 데이터프레임 선언**
---

```python
languages = ['Jullia', 'Python', 'R']
proficiencies = ['30', '10', '20']

tidydata = pd.DataFrame({'language' : languages, 'proficiency' : proficiencies})
tidydata
```

### **B. barplot 산출**
---

```python
fig = ggplot(tidydata)
bar = geom_col(aes(x = 'language', y = 'proficiency', fill = 'language'))

fig + bar
```
